#EXTRACCIÓN DE DATOS DESDE APIs

La API usada en el presente notebook es de la pagina de APIs publicas conocida como Rapidapi. En su sección de API Hub se selecciono un Dataset del historial de datos de la Pandemia del Coronavirus a nivel Mundial. Esta API recibe como request mensajes HTTP tipo GET y envia como respuestas datos en formato JSON. 

URL API: https://rapidapi.com/api-sports/api/covid-193

##TIPOS DE DATOS 

De esta API podemos extraer tres tipos de datos:

1. Lista de Paises donde se registra información sobre la situación de coronavirus. 
2. Estadisticas de muertes y vacunados por pais. 
3. Historial de evolución de la pandemia en cada pais. 

Todos los requests abren una conexión HTTPS, donde se define el URL de la API, adicionalmente, se definen los headers bajo una estructura tipo JSON, donde se indica el API Key y el API Host. Posteriormente se abre una conexión request (socket) donde se define el metodo HTTP, los datos que se desean extraer (Paises, Estadisticas o Historial) y por ultimo los headers. Por ultimo se lee la respuesta recibida de la API.

###Estadisticas

A continuación nos extraemos de la API las estadisticas que contiene de cada pais referente a la pandemia del coronavirus. 

```python
import requests

url = "https://covid-193.p.rapidapi.com/statistics"

headers = {
	"X-RapidAPI-Key": "725f9aec22msh5568fece18d0a31p143f68jsne4de93c29d91",
	"X-RapidAPI-Host": "covid-193.p.rapidapi.com"
}

response_estadistica = requests.request("GET", url, headers=headers)

respuesta_estadistica = response_estadistica.text
json.loads(respuesta_estadistica)
```

```python
#Transformamos nuestro JSON respuesta en un Dataframe
import json
from pandas.io.json import json_normalize
dict = json.loads(respuesta_estadistica)
df2 = json_normalize(dict['response']) 
df2.head(20)
```

```python
#Solicita al usuario el Pais que desea buscar información sobre el Covid-19
country_name = input("Enter username:")
print("Username is: " + country_name)
df2[df2["country"] == country_name]
```

```python
#Podemos observar el numero de muertes por continente 
cont = df2.groupby('continent').sum()
print(cont)
```

```python
#Graficamos el numero de muertes totales por Continente
df = pd.DataFrame({'lab':['Africa', 'Asia', 'Europa', 'Norte America', 'Oceania', 'Sur America'], 'val':[515106, 2953108, 3831610, 3067062, 40958, 2656786]})
ax = df.plot.bar(x='lab', y='val', rot=90, figsize=(20, 5))
ax.set_title("Muertes por Continente")
ax.set_xlabel("Continente")
ax.set_ylabel("Numero de muertes (Millones)")
```

Podemos observar que el continente con mayor numero de muertes por coronavirus es Europa, debido a su población longeva y densidad. El continente con menor numero de muertes registradas es Oceania, debio a que esta conformada por pequeñas islas y la densidad poblacional es muy baja. 

```python
#Numero de casos Activos de coronavirus a nivel mundial (Porcentaje)
df = pd.DataFrame({'Casos Activos Coronavirus': [709272, 9146263, 7766625, 7191759, 287029, 889448]},
                  index=['Africa', 'Asia', 'Europa', 'Norte America', 'Oceania', 'Sur America'])
plot = df.plot.pie(y='Casos Activos Coronavirus', figsize=(10, 10))
```

###Historial de casos de coronavirus por pais

La API nos permite extraer información historial de la pandemia del coronavirus por cada pais. lo cual permite hacer analisis de regresión lineal. Esta extracción se realiza a traves de un request http, donde se define como parametros el pais y la fecha de los datos. 

```python
import requests #Importamos libreria request
import json #Importamos libreria json

url = "https://covid-193.p.rapidapi.com/history" #Definimos URL

#Solicita al usuario el Pais que desea buscar información sobre el Covid-19 y la fecha especifica
country_name = input("Ingresa el nombre del pais que desea ver historial de covid-19 (En Ingles):")
date_name = input("Ingresa la fecha que desa obtener información (AAAA-MM-DD):")
print("El pais introducido es: " + country_name)
print("La fecha introducida es: " + date_name)
#df2[df2["country"] == country_name]

querystring = {"country":country_name,"day":date_name} #Parametro a buscar

headers = {
	"X-RapidAPI-Key": "725f9aec22msh5568fece18d0a31p143f68jsne4de93c29d91", #API Key
	"X-RapidAPI-Host": "covid-193.p.rapidapi.com" #API Host
}

response = requests.request("GET", url, headers=headers, params=querystring) #Contrucción del Request

json.loads(response.text) #Carga de la respuesta tipo JSON
```
